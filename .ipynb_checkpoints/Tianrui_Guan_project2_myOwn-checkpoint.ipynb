```python
import sqlite3
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Part 1  Wrangling
### Problem 1


```python
sqlite_file = 'lahman2014.sqlite'
conn = sqlite3.connect(sqlite_file)
# salaries table
salary_query = "SELECT yearID, teamID, sum(salary) as total_payroll FROM Salaries GROUP BY yearID, teamID"

team_salaries = pd.read_sql(salary_query, conn)
team_salaries.head()
```

```python
# wins table
wins_query = "SELECT yearID, teamID, W as wins, G as games, (W*1.0/G*100) as win_percent FROM Teams"

team_wins = pd.read_sql(wins_query, conn)
team_wins.head()
team_wins.dtypes
```

```python
# Here we check that salaries table and team table do not contain Nan (or null). So we don't have to worry about that.

null_salaries_table = team_salaries[team_salaries.isnull().any(axis=1)]
print(null_salaries_table)

null_wins_table = team_wins[team_wins.isnull().any(axis=1)]
print(null_wins_table)
```

```python
# Notice that we can use inner join query, or use inner join for dataframe. 
# Given that there is no NaN in those two tables:
# There are more rows in the team table, which contains more different years. 
# use inner join on yearID and teamID to eliminate the missing data, because the joined
# table will have yearID-teamID pairs that appear in both of the original table.

# table = pd.merge(team_salaries, team_wins, how='inner', left_on=['yearID', 'teamID'], right_on=['yearID', 'teamID'])

join_query = "SELECT Teams.yearID as yearID, Teams.teamID as teamID, (W*1.0/G*100) as win_pct, total_payroll FROM Teams INNER JOIN (SELECT yearID, teamID, sum(salary) as total_payroll FROM Salaries GROUP BY yearID,teamID) as Total ON Teams.yearID=Total.yearID and Teams.TeamID=Total.TeamID"

table = pd.read_sql(join_query, conn)
# close the database
conn.close()
table.head()
```

Indicate missing data:<br>
1 From the two queries above, we don't find any null data in the columns that we are interested in. <br>
2 There are missing rows for specific year and team, so some rows are unmatched. The Salaries table has 860 rows. The Teams table has 2775 rows. After the inner join, the resulted table has 858 rows. Thus, 2 rows are unmatched in Salaries and 1917 rows are unmatched in Teams.

Used inner join to deal with it, so the result only contains the rows from Teams and Salaries with the same combination of teamID and yearID. It will eliminate the missing data if there is a row with a specific year and team not showing up in both tables.

## Correlation between payroll and winning percentage
### Problem 4

```python
# Gives a name var for the five different groups
disc_var = ['1990-1994', '1995-1999', '2000-2004', '2005-2009', '2010-2014']
table_discrete = pd.DataFrame(table90_14)

# use cut to seperate years and assign rows to five different group labeled with disc_var
table_discrete['group'] = pd.cut(table_discrete['yearID'],5, labels=disc_var)

# Take the means for each groups within each period and rename them as mean_pct and mean_payroll
# Here are two usages for getting the means
# linear = table_discrete[['discrete_year', 'teamID', 'win_pct', 'total_payroll']].groupby(['discrete_year', 'teamID']).mean().dropna()
linear = table_discrete[['group', 'teamID', 'win_pct', 'total_payroll']].groupby(['group', 'teamID']).aggregate('mean').dropna()
linear.rename(columns= {'win_pct':'mean_pct', 'total_payroll':'mean_payroll'}, inplace=True)

# for each period, plot a scatter plot and linear regresion
for period in disc_var:
    # locate the group and plot the whole teams
    temp = linear.loc[period]
    temp.plot.scatter(x='mean_payroll', y='mean_pct', c='Black', title=period, label='teams')
    
    # In preparation for Question 2, we highlight some of the standouts and team OAK, 
    # decided by higher than one standard deviation above mean.
    means = temp.mean()
    stds = temp.std()
    outlier = temp[temp.mean_pct > means.mean_pct + stds.mean_pct]
    # plot standout teams
    for index, row in outlier.iterrows():
        plt.scatter(row['mean_payroll'], row['mean_pct'], label=index, s=10)
    # plot team OAK
    plt.scatter(temp.loc['OAK', 'mean_payroll'], temp.loc['OAK', 'mean_pct'], label='OAK', marker='*', s=150)

    # plot the regression line for all teams 
    xs = np.linspace(temp.mean_payroll.min(), temp.mean_payroll.max())
    line = np.polyfit(temp.mean_payroll, temp.mean_pct, 1)
    plt.plot(xs, np.poly1d(line)(xs), '-')
    
    # legend on the cornor
    plt.legend(loc='lower right')
    
    
```

# Part 3 Data Transformations
## Standardizing across Years
### Problem 5 

```python
table_std = pd.DataFrame(table_discrete)

# extract means and stds for each year's total payroll and rename them as means and stds
means = table_std[['yearID', 'teamID', 'total_payroll']].groupby(['yearID']).mean()
means.rename(columns={'total_payroll':'means'}, inplace=True)
stds = table_std[['yearID', 'teamID', 'total_payroll']].groupby(['yearID']).std()
stds.rename(columns={'total_payroll':'stds'}, inplace=True)

# merge the extracted means and stds and associate them back to yearID, teamID pair
stat = pd.merge(table_std[['yearID', 'teamID', 'total_payroll']], pd.merge(means, stds, how='left', left_index=True, right_index=True), how='left', left_on='yearID', right_on='yearID')

# calculate standardized payroll by the given formula
stat['std_payroll'] = (stat.total_payroll - stat.means)/stat.stds

# join the calculated std_payroll back to the original table
data_std = pd.merge(table_std, stat[['yearID', 'teamID', 'std_payroll']], how='left', left_on=['yearID', 'teamID'], right_on=['yearID', 'teamID'])

# check if there is any none entries
# table_std[table_std.isnull().any(axis=1)]

data_std.head()
```

### Problem 6

```python
# Exactly the same thing did in problem 4, except using std_mean_payroll other than total_mean_payroll
linear_std = data_std[['group', 'teamID', 'win_pct', 'std_payroll']].groupby(['group', 'teamID']).aggregate('mean').dropna()

linear_std.rename(columns= {'win_pct':'mean_pct', 'std_payroll':'std_mean_payroll'}, inplace=True)

for period in disc_var:
    temp = linear_std.loc[period]
    temp.plot.scatter(x='std_mean_payroll', y='mean_pct', c='Black', title=period, label='teams')
    
    means = temp.mean()
    stds = temp.std()
    outlier = temp[temp.mean_pct > means.mean_pct + stds.mean_pct]
    for index, row in outlier.iterrows():
        plt.scatter(row['std_mean_payroll'], row['mean_pct'], label=index, s=10)
    plt.scatter(temp.loc['OAK', 'std_mean_payroll'], temp.loc['OAK', 'mean_pct'], label='OAK', marker='*', s=150)

            
    xs = np.linspace(temp.std_mean_payroll.min(), temp.std_mean_payroll.max())
    line = np.polyfit(temp.std_mean_payroll, temp.mean_pct, 1)
    plt.plot(xs, np.poly1d(line)(xs), '-')
    plt.legend(loc='lower right')
    
```

In problem 4, the domain of mean_payroll(x) was around 10e7 to 2*10e8, but in problem 6, the data points are standardized, which is centered at 0 by its mean (std_mean_payroll sums up to 0), and ranges approximately from -2 to 3.
